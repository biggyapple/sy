```python

df1 = pd.read_csv('card.csv',encoding = 'cp949')
info = df1
info.sort_values(by='년월')
info.drop(['지역구분','업종코드','이용자 구분','관광구분','데이터기준일자','업종명 대분류'],axis=1,inplace=True)
info
#값 없는거 0으로 채움
info2 =  info['연령대'].fillna(0)
info['연령대'] = info2

#카테고리별로 업종 나누기
somae = info[info['업종명'].isin(['과실 및 채소 소매업','관광 민예품','수산물 소매업','육류 소매업','빵 및 과자 소매업','화장품 및 방향제 소매업'])]
food = df1[df1['업종명'].isin(['비알콜 음료점업','피자, 햄버거, 샌드위치 및 유사 음식점업','한식 음식점업','중식 음식점업','일식 음식점업','서양식 음식점',
         '기타 외국식 음식점업','비알콜 음료점업'])]

sleep = df1[df1['업종명'].isin(['휴양콘도 운영업','여관업','호텔업'])]
acti = df1[df1['업종명'].isin(['스포츠 및 레크레이션 용품 임대업','기타 수상오락 서비스업'])]

#이름 통일
somae['업종명'] = '소매업'
food['업종명'] = '식음료'
sleep['업종명'] = '숙박'
acti['업종명'] = '문화레저'

#반복문으로 사용하기 위해 리스트로 만듦
temp = [somae,food,sleep,acti]


li = []
for key in temp:
    
    key = key.sort_values(by='연령대')
    key = key[key.이용금액 < 5000000]
    # somae['연령대'].get_group('0.0')
    index = key[key['성별']=='알수없음'].index
    # print(index)
    key = key.drop(index)

    pd.options.display.float_format = '{:.3f}'.format
    key['연령대']= key[['연령대']].astype(int)
    somae_10_20 = key[(key['연령대']==10) | (key['연령대']==20)]
    somae_30_40_50 = key[(key['연령대']==30) | (key['연령대']==40)| (key['연령대']==50)]
    somae_60_70 = key[(key['연령대']==60) | (key['연령대']==70)]
    # info = somae.groupby('연령대').get_group(10)


    f = somae_10_20.groupby(['읍면동명','업종명']).mean()
    f2 = somae_30_40_50.groupby(['읍면동명','업종명']).mean()
    f3 = somae_60_70.groupby(['읍면동명','업종명']).mean()

    f.drop(['매장수'],axis=1,inplace=True)
    f2.drop(['매장수'],axis=1,inplace=True)
    f3.drop(['매장수'],axis=1,inplace=True)


    
    f['연령대'] = "10_20"
    f2['연령대'] = "30_40_50"
    f3['연령대'] = "60_70"

    f.columns = ["연령대","1인당 이용금액"]
    f2.columns = ["연령대","1인당 이용금액"]
    f3.columns = ["연령대","1인당 이용금액"]
    
    #나온 결과 리스트로 받아서 반복문 끝나고 합침
    result = pd.concat([f,f2,f3])
    li.append(result)
result1 = pd.concat(li)
result1.to_csv('cardone.csv',encoding='cp949')
```
